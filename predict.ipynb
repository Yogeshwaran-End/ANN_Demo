```python
import tensorflow as tf
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
import pandas as pd 
```

```python
#load the tranined model, scaler pickel, onehot
import pickle
model = load_model('model.h5') 

## load the encoder and scaler 
with open('lable_encode_gender.pkl','rb') as file :
    lable_encode_gender =   pickle.load(file)
with open('one_hot_encoder_geo.pkl','rb') as file :
    one_hot_encoder_geo =   pickle.load(file)
with open('scalar.pkl','rb') as file :
    scalar =   pickle.load(file)
```

```python
input_d ={
  'CreditScore': 601,
  'Geography': 'France',
  'Gender': 'Male',
  'Age': 74,
  'Tenure': 2,
  'Balance': 0.0,
  'NumOfProducts': 2,
  'HasCrCard': 0,
  'IsActiveMember': 1,
  'EstimatedSalary': 51554.58,
    }
```

```python
print(type(one_hot_encoder_geo))
input_d = pd.DataFrame([input_d])
```

```python
geo = one_hot_encoder_geo.transform([input_d['Geography']]).toarray()
df = pd.DataFrame(geo,columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
df
```

```python
input_d = pd.concat([input_d.reset_index(drop=True), df], axis=1)
```

```python
input_d['Gender'] = lable_encode_gender.transform(input_d['Gender'])
```

```python
input_d = input_d.drop(columns='Geography')
```

```python
input_scaler =  scalar.transform(input_d)
```

```python
prediction = model.predict(input_scaler)
prediction
```
